# Python Expense Tracker
This notebook contains a simple CLI-based expense tracker that demonstrates the use of lists, functions, and control flow.

```python
def display_menu():
    """Displays the options available to the user."""
    print("\n--- Expense Tracker Menu ---")
    print("1. Add Expense")
    print("2. View Expenses")
    print("3. View Total Expense")
    print("4. Exit")

def add_expense(expenses):
    """Prompts user for input and adds an expense to the list."""
    name = input("Enter the expense name: ")
    try:
        amount = float(input("Enter the amount: "))
        expenses.append({"name": name, "amount": amount})
        print(f"Successfully added: {name}")
    except ValueError:
        print("Invalid input. Please enter a numeric value for the amount.")

def view_expenses(expenses):
    """Prints all recorded expenses."""
    if not expenses:
        print("\nNo expenses recorded yet.")
    else:
        print("\n--- Your Expenses ---")
        for idx, item in enumerate(expenses, 1):
            print(f"{idx}. {item['name']}: ${item['amount']:.2f}")

def view_total(expenses):
    """Calculates and displays the sum of all expenses."""
    total = sum(item['amount'] for item in expenses)
    print(f"\nTotal Expenses: ${total:.2f}")

def main():
    expenses = []
    while True:
        display_menu()
        choice = input("Select an option (1-4): ")

        if choice == '1':
            add_expense(expenses)
        elif choice == '2':
            view_expenses(expenses)
        elif choice == '3':
            view_total(expenses)
        elif choice == '4':
            print("Exiting Expense Tracker. Goodbye!")
            break
        else:
            print("Invalid choice, please try again.")

# Run the program
main()
```
